```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline 

housing = pd.read_csv('Ames_HousePrice.csv', index_col=0)
housing.shape
```

### Highest median prices by neighborhood

```python
plt.figure(figsize=(10,8))
housing.groupby('Neighborhood').SalePrice.median().sort_values().plot.barh()
plt.xlabel('Median Sale Price')
```

### Standard Deviation of Sale Price, by Nieghborhood

```python
plt.figure(figsize=(10,8))
housing.groupby('Neighborhood').SalePrice.std().fillna(0).sort_values(ascending=True).plot.barh()
plt.xlabel('Standard Deviation of Sale Price')
```

### Number of Houses Sold by Neighborhood

```python
plt.figure(figsize=(10,8))
housing.groupby('Neighborhood').SalePrice.count().sort_values().plot.barh()
plt.xlabel('Number of Houses Sold Per Neighborhood')
```

### Higher sale price, greater the standard deviation in the sale

```python
price_med = housing.groupby('Neighborhood').SalePrice.median()
price_sd = housing.groupby('Neighborhood').SalePrice.std().fillna(0)
price_count = housing.groupby('Neighborhood').SalePrice.count()
```

```python
plt.figure(figsize=(10,8))
plt.scatter(price_med,price_sd)
plt.xlabel("Median Sale Price")
plt.ylabel("Standard Deviation in Sale Price")
plt.title("Median Sale Price vs. Sale Price Standard Dev, by Neighborhood")
```

### What are the oldest neighborhoods in Ames?

```python
year_med = housing.groupby('Neighborhood').YearBuilt.median()
year_sd = housing.groupby('Neighborhood').YearBuilt.std().fillna(0)
year_count = housing.groupby('Neighborhood').YearBuilt.count()
```

```python
plt.figure(figsize=(10,8))
year_med.sort_values().plot.barh()
plt.xlim(1900,2020)
plt.xlabel("Median of Year Built")
```

```python
plt.figure(figsize=(10,8))
year_sd.sort_values().plot.barh()
plt.xlabel("Standard Deviation of Year Built")
```

### What is the median sale price over the course of the dataset?

```python
plt.figure(figsize=(12,10))
housing.groupby(['YrSold','MoSold']).SalePrice.median().plot.bar()
plt.ylim(100000,200000)
plt.xlabel("Year and Month Sold")
plt.ylabel("Median sale price")
```

```python
plt.figure(figsize=(12,10))
housing.groupby(['YrSold','MoSold']).SalePrice.std().plot.bar()
plt.xlabel("Year and Month Sold")
plt.ylabel("Standard Deviation of sale price")
```
